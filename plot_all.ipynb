```python
from easygplot import read, Plot
data_obj = read.CSV("gasprices.csv")
```

# Plotting all the graphs with all indexes

```python
g1 = Plot.plot_all(data_obj[:])
```

# Plotting all the graphs with only one index

```python
g2 = Plot.plot_all(data_obj[2])
```

# Plotting all the graphs by slicing the index

```python
g3 = Plot.plot_all(data_obj[3:6])
```
