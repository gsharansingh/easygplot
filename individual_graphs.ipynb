# Read the csv file

```python
from easygplot import read, Plot
data_obj = read.CSV("gasprices.csv")
```

# Method 1: passing data to the object

```python
line_plot = Plot.Line(legend=True)
g1 = line_plot(data_obj[:])
```

# Method 2: passing data while initializing the object

```python
line_plot = Plot.Line(data_obj[:], legend = True)
g2 = line_plot()
```

# Method 3: passing data inline

```python
g3 = Plot.Line(legend=True)(data_obj[:])
```

# Passing single index data

```python
line_plot = Plot.Line(data_obj[2], legend = True)
g4 = line_plot()
```

# Other Plots

## Box Plot

```python
b1 = Plot.Box(title = "Box Plot with 3 indexes")(data_obj[3:6])
b2 = Plot.Box(title = "Box Plot with a single index")(data_obj[2])
```

## Pie Plot

```python
p1 = Plot.Pie(title_cwise = "Pie chart 1: Box Plot with 3 indexes", title_rwise = "Pie chart 2: Box Plot with 3 indexes", legend=True)(data_obj[3:6])
p2 = Plot.Pie(title_cwise = "Box Plot with single index", legend=True)(data_obj[2])
```

## Histogram

```python
h1 = Plot.Histogram(title = "Histogram with 3 indexes", legend=True)(data_obj[3:6])
h2 = Plot.Histogram(title = "Histogram with single index", legend=True)(data_obj[2])
```

## Bar Plot

```python
bar1 = Plot.Bar(title = "Bar Plot with 3 indexes", legend=True)(data_obj[3:6])
bar2 = Plot.Bar(title = "Bar Plot with single index", legend=True)(data_obj[2])
```
